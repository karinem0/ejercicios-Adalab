```python
#Ahora, debemos sacar la informacion que necesitamos de cada uno y almacenarlo en un diccionario que tendrá listas vacías a las que le añadiremos la formación que extraemos. 
diccionario = {'nombres': [], 'apellidos' : [], 'equipos' : []}

#Definimos url y usamos f string para introducir variable pag 

#Creamos bucle:
#Este rango ya que empieza en pagina 1, 2, 3... y son 25 jugadores por pagina (25*8)
for pag in range(1,9):     
    url = f"https://www.balldontlie.io/api/v1/players?page={pag}"
    jugadores = requests.get(url)
    jugadores_json = jugadores.json()
    
    for jugador in jugadores_json['data']:
        nombre = jugador['first_name']
        apellido = jugador['last_name']
        equipos = jugador['team']['full_name']  
        
        diccionario['nombres'].append(nombre)
        diccionario['apellidos'].append(apellido)
        diccionario['equipos'].append(equipos)
    #Nos quedará apendear en las listas vacias del diccionario: 
        
    
    #Con esto hemos logrado que nos saquen los 200 jugadores que necesitamos
    
```

- Extrae información de todos los partidos jugados, en concreto deberás sacar la información del nombre del equipo local, nombre del equipo visitante, puntos del equipo local y del visitante y la temporada a la que pertenece el partido. Para eso tendrás que usar el [siguiente](https://www.balldontlie.io/home.html#get-all-games) endpoint. Extrae la información de 200 partidos. Devuelve los resultados en un DataFrame. Te deberá quedar algo parecido a esto: 

        ![partidos](https://github.com/Adalab/data_imagenes/blob/main/Modulo-2/api's_clase_invertida_partidos.png?raw=true)

```python

# llamada_partido = requests.get(url_partido)
# print(f'La llamada a la API nos ha dado una respuesta de tipo: {llamada_partido.status_code}')

# if llamada_partido.status_code == 200:
#     data_partido = llamada_partido.json()
# else:
#     print("Error al obtener los datos de la API")
```

```python

diccionario_partidos = {'nombres_local' : [], 'nombres_visitante' : [], 'puntos_local' : [], 'puntos_visitante' : [], 'temporada' : []}

for pag in range(1,9):     
    url_partido = f"https://www.balldontlie.io/api/v1/games?pages={pag}"
    llamada_partido = requests.get(url_partido)
    partidos_json = llamada_partido.json()

    for partido in partidos_json['data']: 
        nombre_local = partido['home_team']['full_name']
        nombre_visitante = partido['visitor_team']['full_name']
        puntos_local = partido['home_team_score']
        puntos_visitante = partido['visitor_team_score']
        temporada = partido['season']

        diccionario_partidos['nombres_local'].append(nombre_local)
        diccionario_partidos['nombres_visitante'].append(nombre_visitante)
        diccionario_partidos['puntos_local'].append(puntos_local)
        diccionario_partidos['puntos_visitante'].append(puntos_visitante)
        diccionario_partidos['temporada'].append(temporada)
```
